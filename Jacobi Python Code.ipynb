Pada kesempatan kali ini, kami mencoba menghitung menggunakan Metode Jacobi. Untuk soalnya sebagai berikut : 

22x + 21y - z = -15

3x - 21y + 20z = 25

12x + 2y + 15z = 13





```python
# Mendefiniskan persamaan yang akan dipecahkan dan diinisialisasi ke f1, f2, dan f3
f1 = lambda x,y,z: (-15-21*y+z)/22
f2 = lambda x,y,z: (25-3*x-20*z)/21
f3 = lambda x,y,z: (13-12*x-2*y)/15

# inisialisasi kondisi awal
x0 = 0
y0 = 0
z0 = 0
count = 1

# membaca toleransi error 
e = float(input('Masukkan toleransi error : '))

# Implementasi dari iterasi Jacobi 
print('\nCount\tx\ty\tz\n')

condition = True

while condition:
    x1 = f1(x0,y0,z0)
    y1 = f2(x0,y0,z0)
    z1 = f3(x0,y0,z0)
    print('%d\t%0.4f\t%0.4f\t%0.4f\n' %(count, x1,y1,z1))
    e1 = abs(x0-x1);
    e2 = abs(y0-y1);
    e3 = abs(z0-z1);
    
    count += 1
    x0 = x1
    y0 = y1
    z0 = z1
    
    condition = e1>e and e2>e and e3>e

print('\nSolution: x=%0.3f, y=%0.3f and z = %0.3f\n'% (x1,y1,z1))
```
